```python
#Author: Rahul Kejriwal
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor as gb
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing 
```

```python
train_x = pd.read_csv("../../volume/data/raw/train_file.csv")
test_data = pd.read_csv("../../volume/data/raw/test_file.csv")
train_x = train_x.drop("id", axis = 1)
test_x = test_data.drop("id", axis = 1)

X, y = train_x.iloc[:,:-1],train_x.iloc[:,-1] # Y is the target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)


eval_set = [(X_test, y_test)]
xgb_reg = xgb.XGBRegressor(max_depth=2, booster = "gbtree", grow_policy='depthwise', normalize_type = 'forest', csv_weights=2, learning_rate = 0.005, num_round = 2000, nfold = 5,nrounds = 10,n_estimators=2500,min_child_weight=4, sample_type='weighted',colsample_bytree=0.5,subsample=0.8,eta=0.02,seed=4)
xgb_fit = xgb_reg.fit(X_train, y_train, eval_set = eval_set, eval_metric = 'logloss', early_stopping_rounds =70)


predictions=  xgb_fit.predict(test_x)
test_data["result"] = predictions
result = test_data[["id", "result"]]
result.to_csv("../../volume/data/processed/final.csv", index = False)
```
